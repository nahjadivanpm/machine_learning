```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product_names = ['Teh Pucuk', 'Beng-beng', 'Choki-choki', 'Kopiko', 'Torabika', 'Energen']
product_counts = [52, 39, 78, 13, 78, 29]

df_names = pd.DataFrame(product_names, columns =['names'])
df_stok = pd.DataFrame(product_counts, columns = ['stok'])

df = df_names.assign(stok = df_stok)

print(df)
```

```python
ax = plt.gca()
df.plot(kind='line',x='names',y='stok', color='blue', ax=ax, marker='o')

plt.show()
```

```python
plt.pie(
    product_counts,
    labels=product_names,
    autopct='%1.1f%%',
)
plt.show()
```
